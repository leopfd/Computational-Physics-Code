```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 400
plt.rcParams['figure.figsize'] = [10,2]
```

### 3. Writing an array to a text file

```python
h = float(input('Enter a height (m): '))
v_0 = float(input('Enter an initial velocity (m/s): '))
theta = float(input('Enter an angle (rad): '))
g = 9.8

def projectile(h,v_0,theta):
    x = []
    y = []
    position = [x,y]
    y_val = h
    t = 0
    while y_val >= 0:
        x_val = v_0*t*np.cos(theta)
        y_val = v_0*t*np.sin(theta)-0.5*g*t**2+h
        if y_val >= 0:
            x.append(round(x_val,3))
            y.append(round(y_val,3))
        t += 0.05
    return position

projectile_motion = np.asarray(projectile(h,v_0,theta))
print(projectile_motion)
np.savetxt('projectile.txt',np.transpose(projectile_motion), delimiter=',', header='x,y')
print('file saved')
```

### 7. Plotting user-defined functions

```python
def cos(x):
    return np.cos(x)
def sin(y):
    return np.sin(y)
```

```python
vals = np.linspace(-2*np.pi, 2*np.pi, 200)
cos_vals = cos(vals)
sin_vals = sin(vals)
```

```python
y = [0,1,2]
fig, ax = plt.subplots(1,3, sharex=True, sharey=True)
for z in range(3):
    if z == 0:
        ax[y[z]].plot(vals,cos_vals,c='blue')
        ax[y[z]].set_ylabel('y')
    elif z == 1:
        ax[y[z]].plot(vals,sin_vals,c='red')
    else:
        ax[y[z]].plot(vals,cos_vals,c='blue')
        ax[y[z]].plot(vals,sin_vals,c='red')
    ax[y[z]].tick_params(which='major', direction='in', length=2)
    ax[y[z]].tick_params(which='minor', direction='in')
    ax[y[z]].set_xlim(-2*np.pi,2*np.pi)
    ax[y[z]].set_ylim(-1.1,1.1)
    ax[y[z]].set_xlabel('x')
plt.subplots_adjust(wspace=0.05, hspace=0.05)
plt.show()
```

```python
vals = np.linspace(-2*np.pi, 2*np.pi, 20)
cos_vals = cos(vals)
sin_vals = sin(vals)

y = [0,1,2]
fig, ax = plt.subplots(1,3, sharex=True, sharey=True)
for z in range(3):
    if z == 0:
        ax[y[z]].scatter(vals,cos_vals,marker='*',c='blue')
        ax[y[z]].set_ylabel('y')
    elif z == 1:
        ax[y[z]].scatter(vals,sin_vals,marker='^',c='red')
    else:
        ax[y[z]].plot(vals,cos_vals,'*-',c='blue')
        ax[y[z]].plot(vals,sin_vals,'^--',c='red')
    ax[y[z]].tick_params(which='major', direction='in', length=2)
    ax[y[z]].tick_params(which='minor', direction='in')
    ax[y[z]].set_xlim(-2*np.pi,2*np.pi)
    ax[y[z]].set_ylim(-1.1,1.1)
    ax[y[z]].set_xlabel('x')
plt.subplots_adjust(wspace=0.05, hspace=0.05)
plt.show()
```

### 8. Axis Titles, Legends, and Fonts

```python
vals = np.linspace(-2*np.pi, 2*np.pi, 200)
cos_vals = cos(vals)
sin_vals = sin(vals)

y = [0,1,2]
fig, ax = plt.subplots(1,3, sharex=True, sharey=True)
for z in range(3):
    if z == 0:
        ax[y[z]].plot(vals,cos_vals,c='blue',label='cos(x)')
        ax[y[z]].set_ylabel('y')
    elif z == 1:
        ax[y[z]].plot(vals,sin_vals,c='red',label='sin(x)')
    else:
        ax[y[z]].plot(vals,cos_vals,c='blue',label='cos(x)')
        ax[y[z]].plot(vals,sin_vals,c='red',label='sin(x)')
    ax[y[z]].tick_params(which='major', direction='in', length=2)
    ax[y[z]].set_xlim(-2*np.pi,2*np.pi)
    ax[y[z]].set_ylim(-1.1,1.1)
    ax[y[z]].set_xlabel('x')
plt.suptitle('Sine waves', fontsize=9)
plt.legend(loc='upper right',fontsize=6)
plt.subplots_adjust(wspace=0.05, hspace=0.05)
plt.show()
```

### 11. Random Number generation: Rolling a Die

```python
rolls = [[],[],[],[],[],[]]
allrolls = []
for i in range(1000000):
    die = np.random.default_rng().integers(1,7)
    allrolls.append(die)
    rolls[die-1].append(die)
print(rolls)

plt.hist(allrolls, 6)
plt.show()
```

### 12. Rolling two Dice

```python
num_rolls = 0
double_6 = 0
for i in range(1000000):
    die1 = np.random.default_rng().integers(1,7)
    die2 = np.random.default_rng().integers(1,7)
    if die1 == die2 == 6:
        double_6 += 1
    num_rolls += 1
print(double_6/num_rolls)
```
