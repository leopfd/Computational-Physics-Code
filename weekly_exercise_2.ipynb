### 2. Leap years, precisely

```python
y = int(input('Enter a year: '))
if y%4 == 0:
    if y%100 == 0 and y%400 == 0:
        print(f'{y} is a leap year')
    else:
        print(f'{y} is not a leap year')
else:
    print(f'{y} is not a leap year')
```

```python
G = 6.67e-11
M = 5.97e24
R = 6371000
T = float(input('Enter a period of orbit T: '))
h = np.cbrt((G*M*(T**2))/(4*(np.pi**2)))-R

if h < 0:
    print('Period not allowed')
else:
    print(f'h = {round(h,2)} meters')
```

### 7. Building lists

```python
V = float(input('Enter reading (<0 to stop): '))
voltages = []
while V > 0:
    voltages.append(V)
    V = float(input('Enter reading (<0 to stop): '))
print(voltages)
```

### 8. Calculating list statistics

```python
V = float(input('Enter reading (<0 to stop): '))
voltages = []
while V > 0:
    voltages.append(V)
    V = float(input('Enter reading (<0 to stop): '))
voltages = np.asarray(voltages)
print(voltages)
print(f'Readings: {voltages}')
print(f'Number of readings: {len(voltages)}')
print(f'Average reading: {np.average(voltages)}')
print(f'Smallest reading: {min(voltages)}')
print(f'Largest reading: {max(voltages)}')
```

### 12. Nested for statements: atomic orbitals

```python
N = int(input('Enter a principal quantum number (integer): '))
n_numbers = np.arange(1,N+1,1)
for n in n_numbers:
    qstates = []
    l_numbers = np.arange(0,n,1)
    for l in l_numbers:
        if l == 0:
            m_numbers = np.asarray([0])
            for m in m_numbers:
                state = [n,l,m]
                qstates.append(state)
                print(state)
        else:
            m_numbers = np.arange(-l,l+1,1)
            for m in m_numbers:
                state = [n,l,m]
                qstates.append(state)
                print(state)
    if n == 1:
        print(f'There is {len(qstates)} quantum states for principal quantum number {n}')
    else:
        print(f'There are {len(qstates)} quantum states for principal quantum number {n}')
```

### 13. The binomial coefficient

```python
def binomial(n, k):
    p = 1    
    for i in range(1, min(k, n - k) + 1):
        p *= n
        p //= i
        n -= 1
    return p

binomial(10,3)
```
