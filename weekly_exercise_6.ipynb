```python
import numpy as np
from numpy import sqrt,exp
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
from scipy import constants as c
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
data = np.loadtxt('data_handouts/peak4.txt', float,skiprows=1) # read into 2D array 
x = data[:,0]
y = data[:,1]
n = len(x) # number of data points

def f(x,a,x0,w,b0,b1):
    return a*exp(-(x-x0)**2/(2*w**2)) + b0 + b1*x

num_par = 5
guesses = [3,5,1,1,0]

(a,x0,w,b0,b1),c = curve_fit(f,x,y,p0=guesses)
(a_err,x0_err,w_err,b0_err,b1_err) = 2*sqrt(np.diag(c))
yfit = f(x,a,x0,w,b0,b1) 
sq_res = (yfit-y)**2
chisqr = sum(sq_res)/(n-num_par)

# plot data and model together
xmod = np.linspace(x[0],x[-1],100)
ymod = f(xmod,a,x0,w,b0,b1) 
plt.figure(figsize=(7.5,5))
plt.rc('font', size=16) 
plt.plot(x,y,'bo') # plot the data 
plt.plot(xmod,ymod,'r') 
plt.xlabel('x (cm)')
plt.ylabel('y (mV)') 
plt.show()
# report results
# compute reduced chi-squared
# assumes x's are in order
# plot the model
print('reduced chi-squared =',chisqr)
print('parameters and 2-sigma parameter uncertainties:') 
print('A = %f +/- %f'%(a,a_err))
print('x0 = %f +/- %f'%(x0,x0_err))
```

### 5. Radioactive decay - fitting the data

```python
def decay(t, N0, tau, B):
    return N0*np.exp(-t/tau)+B

n = [len(t[0]),len(t[1])]
num_params = 3

yfits = []
chisqr = []
guess = [[100,55,20],[200,55,20]]

for i in range(2):
    (N0, tau, B),cov = curve_fit(f=decay, xdata=t[i], ydata=c[i], sigma=c_err[i], p0=guess[i])
    (N0_err,tau_err,B_err) = 2*sqrt(np.diag(cov))
    yfits.append(decay(t[i], N0, tau, B))
    sq_res = (yfits[-1]-c[i])**2
    chisqr.append(sum(sq_res)/(n[i]-num_par))
    print((N0, tau, B))
print('X^2:',chisqr)

plt.errorbar(t[0], c[0], c_err[0], c='b', fmt='.')
plt.errorbar(t[1], c[1], c_err[1], c='r', fmt='.')
plt.plot(t[0], yfits[0], c='b', label='decay 1')
plt.plot(t[1], yfits[1], c='r', label='decay 2')
plt.xlabel('Time (min)')
plt.ylabel('Counts/min')
plt.legend()
plt.show()
```

### 6. Gaussian vs Lorentzian Fits

```python
def gauss(x,A,x0,w,B):
    return A*exp(-(x-x0)**2/2*w**2)+B

def lorentz(x,A,x0,w,B):
    return A/(1+((x-x0)/w)**2)+B
```

```python
data1 = np.loadtxt('data_handouts/gl1.txt', skiprows=1).transpose()
data2 = np.loadtxt('data_handouts/gl2.txt', skiprows=1).transpose()

x = [data1[0], data2[0]]
y = [data1[1], data2[1]]

n = [len(x[0]),len(x[1])]
num_params = 3

gauss_fits = []
lorentz_fits = []
gauss_guess = [[5,1,1,5],[1,1,1,1]]
lorentz_guess = [[5,1,1,5],[1,1,1,1]]

for i in range(2):
    (A,x0,w,B),cov = curve_fit(f=gauss, xdata=x[i], ydata=y[i], p0=gauss_guess[i])
    (A_err,x0_err,w_err,B_err) = 2*sqrt(np.diag(cov))
    gauss_fits.append(gauss(x[i], A,x0,w,B))
    sq_res = (gauss_fits[-1]-y[i])**2
    gauss_chisqr = (sum(sq_res)/(n[i]-num_par))
    print(f'Gauss {i+1} Parameters: {(A,x0,w,B)}')
    print(f'Gauss {i+1} Errors: {A_err,x0_err,w_err,B_err}')
    print(f'Gauss {i+1} X^2: {gauss_chisqr} \n')

    (A,x0,w,B),cov = curve_fit(f=lorentz, xdata=x[i], ydata=y[i], p0=lorentz_guess[i])
    (A_err,x0_err,w_err,B_err) = 2*sqrt(np.diag(cov))
    lorentz_fits.append(lorentz(x[i], A,x0,w,B))
    lorentz_errors.append((A_err,x0_err,w_err,B_err))
    sq_res = (lorentz_fits[-1]-y[i])**2
    lorentz_chisqr = (sum(sq_res)/(n[i]-num_par))
    print(f'Lorentz {i+1} Parameters: {(A,x0,w,B)}')
    print(f'Lorentz {i+1} Errors: {A_err,x0_err,w_err,B_err}')
    print(f'Lorentz {i+1} X^2: {lorentz_chisqr} \n')

fig, ax = plt.subplots(1,2, figsize=(12, 6), sharex=True, sharey=True)

ax[0].scatter(x[0], y[0], c='b')
ax[0].scatter(x[1], y[1], c='r')
ax[0].plot(x[0], gauss_fits[0], c='b', label='Gauss 1')
ax[0].plot(x[1], gauss_fits[1], c='r', label='Gauss 2')
ax[1].scatter(x[0], y[0], c='b', marker='x')
ax[1].scatter(x[1], y[1], c='r', marker='x')
ax[1].plot(x[0], lorentz_fits[0], c='b', label='Lorentz 1')
ax[1].plot(x[1], lorentz_fits[1], c='r', label='Lorentz 2')
ax[0].set_xlabel('Time (min)')
ax[1].set_xlabel('Time (min)')
ax[0].set_ylabel('Counts/min')
ax[0].tick_params(axis='both',direction='in')
ax[1].tick_params(axis='both',direction='in')
ax[0].legend()
ax[1].legend()
plt.subplots_adjust(wspace=0.05)
plt.show()
```

### 8. Make a 3D surface plot: atomic orbital

```python
x = np.linspace(-20,20,2000)               # row vector of x's
y = np.linspace(-20,20,2000)[:,np.newaxis] # column vector of y's
r = np.sqrt(x**2+y**2)
phi = x*(6-r)*exp(-r/3)

# make the surface plot
plt.figure(figsize=(12,8))   # call the figure fig
plt.rc('font', size=12)
ax = plt.axes(projection='3d')   # get some 3D axes
ax.plot_surface(x,y,phi,cmap=cm.seismic_r)  # do the surface plot
ax.set_zlim(-4., 4.)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.title (r'Surface Plot of $\phi(x,y) = x(6-r)e^{-r/3}$')

plt.show()
```
