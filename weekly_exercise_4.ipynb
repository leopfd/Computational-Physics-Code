```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams['figure.dpi'] = 100
```

### 1. Creating your own function: Cartesian to Polar coordinates

```python
def polar(x,y):
    r = np.sqrt(x**2+y**2)
    theta = np.arctan(y/x)*(180/np.pi)
    return r,theta

for i in range(5):
    x = float(input('Enter an x value: '))
    while x == 0:
        x = float(input('Enter an non-zero x value: '))
    y = float(input('Enter an y value: '))
    print(f'r = {round(polar(x,y)[0],3)}')
    print(f'theta = {round(polar(x,y)[1],3)}')
```

### 3. Parametric/polar plots of a curve

```python
theta = np.linspace(0,2*np.pi,1000)
x = 2*np.cos(theta)+np.cos(2*theta)
y = 2*np.sin(theta)+np.sin(2*theta)

plt.tick_params(which='major', direction='in', length=2)
plt.ylabel('y')
plt.xlabel('x')
plt.plot(x,y,'b')
plt.show()
```

### 4. Multiple plots in one figure

```python
def cos(x):
    return np.cos(x)
def sin(y):
    return np.sin(y)
def tan(n):
    return np.tan(n)
def sec(m):
    return 1/np.cos(m)

vals = np.linspace(-2*np.pi, 2*np.pi, 200)
cos_vals = cos(vals)
sin_vals = sin(vals)
tan_vals = tan(vals)
sec_vals = sec(vals)



x = [0,0,1,1]
y = [0,1,0,1]

fig, ax = plt.subplots(2, 2)#, sharex=True, sharey=True)

for z in range(4):
    ax[x[z],y[z]].tick_params(which='major', direction='in', length=2)
    ax[x[z],y[z]].set_xlim(-2*np.pi,2*np.pi)
    ax[x[z],y[z]].set_ylim(-1.1,1.1)
    ax[x[z],y[z]].set_ylabel('y')
    ax[x[z],y[z]].set_xlabel('x')
    if z == 0:
        ax[x[z],y[z]].plot(vals,cos_vals,c='blue')
        ax[x[z],y[z]].set_title('cosine')
    elif z == 1:
        ax[x[z],y[z]].plot(vals,sin_vals,c='red')
        ax[x[z],y[z]].set_title('sine')
    elif z == 2:
        ax[x[z],y[z]].plot(vals,tan_vals,c='green')
        ax[x[z],y[z]].set_title('tangent')
        ax[x[z],y[z]].set_ylim(-5,5)
    elif z == 3:
        ax[x[z],y[z]].plot(vals,sec_vals,c='purple')
        ax[x[z],y[z]].set_title('secant')
        ax[x[z],y[z]].set_ylim(-5,5)
plt.subplots_adjust(wspace=0.18, hspace=0.18)
plt.show()
```

### 6. Making density plots using imshow()

```python
def color(x,y):
    r = np.sqrt(x**2+y**2)
    return np.sin(r)

D = 50

x = np.linspace(-D,D,1000)
y = np.linspace(-D,D,1000)

X, Y = np.meshgrid(x, y)

plt.imshow(color(X,Y),extent=[-D,D,-D,D],cmap='plasma')
plt.colorbar()
plt.show()
```

```python
def color(x,y):
    r = np.sqrt(x**2+y**2)
    theta = np.arctan2(y,x)
    return np.sin(r+theta)

D = 50

x = np.linspace(-D,D,10000)
y = np.linspace(-D,D,10000)

X, Y = np.meshgrid(x, y)

plt.imshow(color(X,Y),extent=[-D,D,-D,D],cmap='plasma')
plt.colorbar()
plt.show()
```

### 10. Using and re-using a function

```python
def f(x):
    return x+2*np.cos(x**2)

x = np.linspace(-10,10,1000)
y = f(x)

plt.tick_params(which='major', direction='in', length=2)
plt.ylabel('y')
plt.xlabel('x')
plt.plot(x,y,c='b')
plt.show()
```

```python
w = f(x)+10
z = f(x+5)

plt.tick_params(which='major', direction='in', length=2)
plt.ylabel('y')
plt.xlabel('x')
plt.plot(x,w,c='r', label = 'f(x)+10')
plt.plot(x,y,c='b', label = 'f(x)')
plt.plot(x,z,c='g', label = 'f(x+5)')

plt.legend()
plt.show()
```

```python
def f_abs(x):
    return abs(x)+2*np.cos(x**2)

w = f_abs(x)+10
y = f_abs(x)
z = f_abs(x+5)

plt.tick_params(which='major', direction='in', length=2)
plt.ylabel('y')
plt.xlabel('x')
plt.plot(x,w,c='r')
plt.plot(x,y,c='b')
plt.plot(x,z,c='g')
plt.show()
```

### 11. Make a histogram

```python
def f(x):
    return np.cos(x)**2+5

x = np.linspace(0,10,1000)
y = f(x)

plt.tick_params(which='major', direction='in', length=2)
plt.ylabel('y')
plt.xlabel('x')
plt.plot(x,y,c='b')
plt.show()

n = np.linspace(0,1000,100000)
z = f(n)

plt.figure()
plt.hist(z,50)
plt.show
```
