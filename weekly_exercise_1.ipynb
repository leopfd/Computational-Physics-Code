### 9. The factorial function

```python
def factorial(n):
    n_array = np.linspace(1,n,int(n))
    f = 1
    for i in range(1,int(n)):
        f = f*n_array[i]
    return(f)
    
n = float(input('Enter a number: '))

while n%1!=0 or n<0:
    n = float(input('Please enter a positive integer: '))

print(f'{n}! = {factorial(n)}')
```

### 12. Find the precision of floats

```python
x = float(input('Enter a number: '))
e = float(input('Enter a smaller number: '))

while x+e!=x:
    e = float(input('Enter an even smaller number: '))

print(x+e)
```
